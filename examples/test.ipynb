```python
import urllib, urllib.request
import urllib
import time
import feedparser
from termcolor import colored as c
from dateutil.parser import *
from datetime import *
from json2html import *
import json

query = 'rf+sensing'
start = 0
total_results = 40
results_per_iteration = 40
max_results = 20

for i in range(start, total_results, results_per_iteration):

    url = f'http://export.arxiv.org/api/query' \
          f'?search_query={query}' \
          f'&start={start}&max_results={results_per_iteration}' \
          f'&sortBy=lastUpdatedDate&sortOrder=descending'

    response = urllib.request.urlopen(url).read().decode('utf-8')

    start += results_per_iteration
    feed = feedparser.parse(response)
    total_results = len(feed.entries)

    print(f"Results {i} - {i + results_per_iteration}")

    file_data = json2html.convert(json=feed.entries, encode=True)
    open('temp.html', "wt").write(file_data.decode('utf-8'))

    # Run through each entry, and print out information
    for entry in feed.entries:
        print(c('------' * 10, 'white', 'on_yellow'))
        print(c(f'Title:  {entry.title}', 'red'))
        print(c(f'id: {entry.id}', 'green'))
        print(c(f"arxiv-id: {entry.id.split('/abs/')[-1]}", 'green'))
        print(c(f'First Author:  {entry.author}', 'green'))
        updated = parse(entry.updated[0:10]).strftime("%A %d. %B %Y")
        print(c(f'updated:  {updated}', 'green'))
        published = parse(entry.published[0:10]).strftime("%A %d. %B %Y")
        print(c(f'published:  {published}', 'green'))
        print(c(f'Summary:\r\n\r\n{entry.summary}', 'black'))

    # Remember to play nice and sleep a bit before you call
    # the api again!
    #time.sleep(1)
```
